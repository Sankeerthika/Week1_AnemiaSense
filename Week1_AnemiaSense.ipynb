```python
# Week1_AnemiaSense.ipynb
# ------------------------
# Week 1: Data Collection & Preparation

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/anemia.csv")

# Display first few rows
print("üìÑ Dataset Preview:")
display(df.head())

# Check dataset shape
print("\nüßÆ Number of Rows and Columns:", df.shape)

# Check for missing values
print("\nüîç Missing Values:")
print(df.isnull().sum())

# Data types information
print("\n‚ÑπÔ∏è Dataset Information:")
print(df.info())

# Summary statistics
print("\nüìä Summary Statistics:")
display(df.describe())

# Check class balance (Result column)
print("\n‚öñÔ∏è Class Distribution:")
print(df['Result'].value_counts())

# ----------------------------
# Visual Analysis
# ----------------------------
plt.figure(figsize=(12, 5))

# Histogram of Hemoglobin levels
plt.subplot(1, 2, 1)
sns.histplot(df['Hemoglobin'], bins=10, kde=True, color='skyblue')
plt.title("Distribution of Hemoglobin Levels")

# Countplot of Anemia Result
plt.subplot(1, 2, 2)
sns.countplot(x='Result', data=df, palette='viridis')
plt.title("Count of Anemia Cases (0=No, 1=Yes)")

plt.tight_layout()
plt.show()

# Correlation heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Features")
plt.show()

print("\n‚úÖ Data Exploration Complete! Ready for Week 2 (Model Building).")
```
